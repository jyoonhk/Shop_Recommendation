```python
url = 'https://www.fredperry.com/men/polo-shirts'
headers = {"User-Agent":"Mozilla/5.0"}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text,'html.parser')
df = pd.DataFrame(columns = ['Brand','Category', 'Name', 'Price', 'Image URL'])

images = soup.find_all('img')
for image in images:
    id_no = len(df)
    brand = 'Fred Perry'
    category = soup.find(class_ ="page-title").text.strip('\n')
    price = soup.find(class_='price').text.strip('\n')
    link = image['src']
    name = image['alt']
    Gender = 'Male'
    print(brand)
    print(category)
    print(link)
    print(name)
    print(price)
    df1 = pd.DataFrame({
    'ID': int(id_no),
    'Brand':(brand),
    'Gender':(Gender),
    'Category':str(category),
    'Name':[name], 
    'Price':[price], 
    'Image URL':[link]
    })
    df = df.append(df1)
```
