```python

uniqlo = {
    'Mens Casual Outers': 'https://www.uniqlo.com.hk/en_GB/casual_water_repellent_men.html',
    'Mens Blazer_Cardigan_Coats':'https://www.uniqlo.com.hk/en_GB/jackets_coats_men.html',
    'Mens Shorts':'https://www.uniqlo.com.hk/en_GB/shorts_men.html',
    'Mens Jeans':'https://www.uniqlo.com.hk/en_GB/24h_jeans_men.html',
    'Mens Casual Pants': 'https://www.uniqlo.com.hk/en_GB/long_pants_men.html',
    'Mens Trousers': 'https://www.uniqlo.com.hk/en_GB/kando_pants_men.html',
    'Mens T-shirts_tanktops': 'https://www.uniqlo.com.hk/en_GB/short_sleeves_men.html',
    'Mens Polo-shirt':'https://www.uniqlo.com.hk/en_GB/polo_men.html',
    'Mens Longsleeves':'https://www.uniqlo.com.hk/en_GB/long_sleeves_men.html',
    'Mens Casual-shirts': 'https://www.uniqlo.com.hk/en_GB/casual_shirt_men.html',
    'Mens Formal-shirts': 'https://www.uniqlo.com.hk/en_GB/formal_shirt_men.html',

    'Womens Casual Outers': 'https://www.uniqlo.com.hk/en_GB/casual_water_repellent_women.html',
    'Womens Blazer_Cardigan_Coats':'https://www.uniqlo.com.hk/en_GB/jackets_knit_coats_women.html',
    'Womens Shorts':'https://www.uniqlo.com.hk/en_GB/shorts_women.html',
    'Womens Jeans':'https://www.uniqlo.com.hk/en_GB/24h_jeans_women.html',
    'Womens Skirts': 'https://www.uniqlo.com.hk/en_GB/skirts_women.html',
    'Womens Shirts_Blouses': 'https://www.uniqlo.com.hk/en_GB/shirt_women.html',
    'Womens T-shirt_tanktops':'https://www.uniqlo.com.hk/en_GB/short_sleeves_women.html',
    'Womens Longsleeves':'https://www.uniqlo.com.hk/en_GB/long_sleeves_women.html',
    'Womens Sweatwears': 'https://www.uniqlo.com.hk/en_GB/sweat_women.html'
    }

headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.85 Safari/537.36'
}
#Set destination folder to save subfolder of images
#Additional sub folders for each clothing category will be automatically created to save images
folder = 'C:/Users/Yoon Hwan Kim/Documents/Shop_Recommendation_Project/Uniqlo'

shop = "Uniqlo"
#Create empty df to save all information to
df = pd.DataFrame(columns = ['ID', 'Shop', 'Clothing Category','Brand', 'Price', 'Image URL'])

#Loop through links in the dictionary 
for category in uniqlo.keys():
    #Create a new folder for each clothing category, and change to this folder
    try:
        os.mkdir(os.path.join(folder, str(category)))
        os.chdir(os.path.join(folder, str(category)))
    except:
        pass
    #Use Selenium to begin scrape
    url = uniqlo[category]
    driver = webdriver.Chrome()
    driver.get(url)

    time.sleep(8)
    #Scroll down and rescrape, stop when last_height = new_height, since no new images
    total_height = int(driver.execute_script("return document.body.scrollHeight"))
    for i in range(1, total_height, 1):
        driver.execute_script("window.scrollTo(0, {});".format(i))

    time.sleep(4)
    # except TimeoutException

    #Begin scrape once we have scrolled to the bottom of the page
    id_no = len(df)

    # Define the variables 'brand', 'product_category', 'price' and 'url / src' locations in the webpage
    # 'brand' = defined by shop to categorise groups of similar products, e.g. "Mens Sports Performance Training", 'HeatTech catalogue', 'Dry-Fit', 'ColdTech', 'Nike Yoga' etc
    # 'product category' = defined by shop for product types, e.g. 'long sleeved shirt', 'Men's Perfomance shorts', 'Women's Running Tights', 'Men's Basketball Hoodie' 
    # 'src' = url for link to the picture/image. This is used by the requests.get() function to download the image, check there is a 'http:' at the start of the string
    subhtml = driver.page_source
    soup = bs(subhtml, "html.parser")

    for product in soup.find_all(class_="product-content"): # change 'class_ = ...' to how the each image / product is split in the website 
        try:
            brand = product.find_all('p')[1].text
            price = product.find(class_='h-currency h-cny bold').text
            src = product.find(class_='picture-img')['src']

            # Create a dummy dataframe and append to main df
            df1 = pd.DataFrame({'ID': int(id_no),
                'Shop':str(shop),
                'Clothing Category':str(category),
                'Brand':[brand],
                'Price':[price], 
                'Image URL':[src]
                })
            df = df.append(df1)
            
            #Save image into folder
            image_name = str(shop) + str(" ") + str(id_no) + str(" ") + str(brand) + str(".jpg")
            image_link = src
            with open(str(image_name),'wb') as f:
                im = requests.get(image_link, headers= headers)
                f.write(im.content)
            
            
            id_no+=1
        #some will fail... this is expected
        except:
            print('Failed')
            id_no+=1
            pass
```
