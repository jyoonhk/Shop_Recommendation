```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("csv/User_db.csv")
```

```python
df['Category'].value_counts()
```

```python
gen_cat = df.groupby('Category')['Gender'].value_counts()
```

```python
gen_cat.plot.bar(stacked=True)
```
